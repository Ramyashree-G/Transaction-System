```python
con.execute("""CREATE TABLE IF NOT EXISTS transactions (
    transaction_id INTEGER PRIMARY KEY AUTOINCREMENT,
    senders_id TEXT NOT NULL,
    receivers_id TEXT NOT NULL,
    amount INTEGER CHECK(amount > 0),
    date TEXT DEFAULT (DATE(CURRENT_TIMESTAMP)),
    time TEXT,
    category_id TEXT,
    FOREIGN KEY (senders_id) REFERENCES Users(username) ON DELETE RESTRICT,
    FOREIGN KEY (receivers_id) REFERENCES Users(username) ON DELETE RESTRICT,
    FOREIGN KEY (category_id) REFERENCES Category(category_id)
);
""")

con.commit()


#Creating transactions table
#Transactions table is the child table of users and category tables.
```

```python
con.execute("""
CREATE TRIGGER if not exists insert_timestamp
AFTER INSERT ON transactions
BEGIN
    UPDATE transactions
    SET date = DATE('now', 'localtime'), time = TIME('now', '+05:30')
    WHERE rowid = NEW.rowid;
END;
""")
con.commit()
#Triggered after inserting data into transactions table.
#Updating the date in the transactions table to the time the transaction is done
#current timestamp provides time and date, having date(timestamp) returns only the current date, same for time
#fk connects a column in the given table to another tables pk, this means u can access cols of the other connected table
#on insert restrict allows us to restrict change/insertion/deletion in the child table that violates the foreign key constraint
#a trigger works during crud ops, here the triggers r raised if the username in transactions is not present in users username
#tried using on insert restrict and DEFERRABLE INITIALLY DEFERRED didnt work therefore chose a trigger
#raise(abort) raises an error saying invalid id, of type abort(which halts the current sql statement and rollsback to the previous state,
#SELECT 1 is to check for the existence of a row that meets certain conditions but you don't need to retrieve any actual data from the table. It's a way to check whether a condition is true without incurring the overhead of retrieving any specific columns.
```

```python
con.execute("""CREATE TRIGGER if not exists update_balance
AFTER INSERT ON transactions
BEGIN
    UPDATE balance
    SET bankbalance = bankbalance - NEW.amount
    WHERE username = NEW.senders_id;

    UPDATE balance
    SET bankbalance = bankbalance + NEW.amount
    WHERE username = NEW.receivers_id;
END;
""")
#Triggered after data is inserted into transactions table
#Updating the sender's bank balance by deducting the transaction amount and adding the same transaction amount is added to the reciever's bankbalance.
```

```python
cursor = con.cursor()

# Define a function to check if budget exceeded
def check_budget(sender_id, category_id, amount):
    # Fetch budget amount for the sender and category
    cursor.execute("SELECT budget_amount FROM Budget WHERE username = ? AND category_id = ?", (sender_id, category_id))
    budget_amount = cursor.fetchone()
    if budget_amount and amount > budget_amount[0]:
        print(budget_amount)
        print(amount)
        return True  # Budget exceeded
    else:
        return False  # Budget not exceeded

# Define a function to handle the trigger logic
def check_budget_trigger(sender_id, receiver_id, amount):
    # Check if receiver_id is in the predefined list
    if receiver_id in receiver_categories:
        rec_category = receiver_categories[receiver_id]
        # Check if the sender has set a budget for this category and if the transaction amount exceeds it
        if check_budget(sender_id, rec_category, amount):
            # Handle budget exceeded scenario (you can log or take any other action here)
            # For simplicity, I'm just printing a message
            print("Budget Exceeded!")
            return False  # Budget exceeded, don't insert into transactions table
    return True  # Budget not exceeded, proceed with insertion


    #Checks whether the transaction amount is greater and exceed the budget amount with particular category
    #It also checks whether receiver belongs to an organiztion

```

```python
def update_budget(sender_id,reciver_id,amount):
    category_id = receiver_categories[reciver_id]
    x =con.execute(f"select budget_amount from Budget where username  = '{sender_id}' and  category_id = '{category_id}'").fetchone()[0]
    x-=amount
    con.execute(f"update Budget set budget_amount = {x} where username = '{sender_id}' and category_id = '{category_id}'")
    y  =con.execute(f"select total_amount from Budget where username ='{sender_id}' and category_id = '{category_id}'").fetchone()[0]
    y+=amount
    con.execute(f"update Budget set total_amount = {y} where username ='{sender_id}' and category_id = '{category_id}'")

    con.commit()
    print("Transaction inserted successfully!")

    #When the transaction is done within the budget amount the budget table is updated.
    #The budget amount is deducted and added to the total spent amount
```

```python
def authenticate_user(username, pin):
    user = con.execute("SELECT username FROM Users WHERE username=? AND (pin=? OR pin IS NULL)", (username,pin)).fetchone()
    if user:
        return user[0]
    else:
        return None

#Checking whether the user with the entered username exists.
#if exists, authentication is complete. If not it is an invalid/non-existing user.
```

```python
from openpyxl import Workbook, load_workbook  #For excel files
import sqlite3
from datetime import datetime, timedelta  #For date and time

# Function to fetch all updates from SQLite database for each table
def fetch_updates_from_database():
    connection = sqlite3.connect('Team3/dbmsproj.sqlite')
    cursor = connection.cursor()

    updates = {}

    cursor.execute("SELECT * FROM users")
    updates['users'] = cursor.fetchall()

    cursor.execute("SELECT * FROM category")
    updates['category'] = cursor.fetchall()

    cursor.execute("SELECT * FROM budget")
    updates['budget'] = cursor.fetchall()

    cursor.execute("SELECT * FROM balance")
    updates['balance'] = cursor.fetchall()

    cursor.execute("SELECT * FROM transactions")
    updates['transactions'] = cursor.fetchall()

    return updates

# Function to update existing records in Excel file with new data
def update_excel_file(updates):
    wb = Workbook()

    # Update data for each sheet
    for sheet_name, data in updates.items():
        ws = wb.create_sheet(title=sheet_name)

        # Specify custom column names for each sheet
        custom_column_names = {
            'users': ['Username', 'Name', 'Email', 'Encrypted Pin'],
            'category': ['Category ID', 'Category Name'],
            'budget': ['Budget ID', 'Usename', 'Category ID', 'Budget amount','Total Spent', 'Start Date', 'End Date'],
            'balance': ['Balance ID', 'User ID', 'Amount'],
            'transactions': ['Transaction ID', 'Sender Username', 'Receiver Username', 'Amount', 'Date', 'Time', 'Category_ID']
        }

        # Add custom column names to the sheet
        column_names = custom_column_names.get(sheet_name, [])
        for col, name in enumerate(column_names, start=1):
            ws.cell(row=1, column=col, value=name)

        # Add data to the sheet
        for row_idx, row_data in enumerate(data, start=2):
            for col_idx, value in enumerate(row_data, start=1):
                ws.cell(row=row_idx, column=col_idx, value=value)
    wb.remove(wb['Sheet'])  # Remove default sheet
    wb.save('Team3/Records.xlsx')

# Main function to fetch updates and update Excel file
def main():
    updates = fetch_updates_from_database()
    if updates:
        update_excel_file(updates)
        utc_now = datetime.utcnow()  # Get current UTC time
        ist_now = utc_now + timedelta(hours=5, minutes=30)  # Add 5 hours and 30 minutes for IST
        print(f"Excel file updated with new updates at {ist_now.strftime('%Y-%m-%d %H:%M:%S')} IST")
    else:
        print("No updates found.")

if __name__ == "__main__":
    main()
```
